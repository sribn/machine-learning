## Import & Explore Data

```python
df = pd.read_csv('boston_housing.csv')
df
```

### Data Desc

**crim**
per capita crime rate by town.

**zn**
proportion of residential land zoned for lots over 25,000 sq.ft.

**indus**
proportion of non-retail business acres per town.

**chas**
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

**nox**
nitrogen oxides concentration (parts per 10 million).

**rm**
average number of rooms per dwelling.

**age**
proportion of owner-occupied units built prior to 1940.

**dis**
weighted mean of distances to five Boston employment centres.

**rad**
index of accessibility to radial highways.

**tax**
full-value property-tax rate per \$10,000.

**ptratio**
pupil-teacher ratio by town.

**black**
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

**lstat**
lower status of the population (percent).

**medv**
median value of owner-occupied homes in \$1000s.

## Data Viz

### Correlation Map

```python
fig, ax = plt.subplots(1, 2, gridspec_kw={'width_ratios': [5, .5]}, figsize=(14,9))
fig.suptitle('Correlation Heatmap')

# plot all features correlation
p1 = sns.heatmap(ax = ax[0], data = df.corr(), annot = True) 
bot, top = p1.get_ylim()
p1.set_ylim(bot+0.5, top-0.5)
ax[0].set_title('features correlation')

# plot features correlation with medv
p2 = sns.heatmap(ax = ax[1], data = df.corr()[['medv']], annot = True) 
bot, top = p2.get_ylim()
p2.set_ylim(bot+0.5, top-0.5)
ax[1].set_title('medv correlation')


plt.show()
```

### Filtering features with high correlation

```python
medv_corr = df.corr().loc['medv']

# filtering threshold is 0.45, both + and -
f = medv_corr.loc[(medv_corr != 1) & (medv_corr >= .45) | (medv_corr <= -0.45)] 

# saving filtered features as a list
f_ft = list(f.index) 

f
```

### Scatter Plot

```python
fig = plt.figure(figsize=(15,10))

# scatter plot all filtered features vs medv
for i in range(len(f_ft)):    
    ax = fig.add_subplot(231+i)
    ax.scatter(df[f_ft[i]], df['medv'])
    ax.set_title(f_ft[i]+' vs medv')

plt.show()
```

## Data Preprocessing 1

### Using whole dataset

```python
import sklearn
```

```python
df_X = df.drop(['medv'], axis=1)
df_y = df[['medv']]
```

```python
# splitting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.3, random_state=33)
```

```python
# scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)
X_train = scaler.transform(X_train)

scaler.fit(X_test)
X_test = scaler.transform(X_test)
```

## Data Modeling 1

```python
# modelling
from sklearn.linear_model import LinearRegression

boston_model = LinearRegression()
boston_model.fit(X_train, y_train)
```

```python
boston_model.score(X = X_train, y = y_train)
```

### Model Test

```python
pred = boston_model.predict(X_test)
```

```python
# test
from sklearn.metrics import r2_score

det = r2_score(y_test, pred)
det
```

```python
plt.scatter(y_test, pred, c='r')
plt.title('Model 1')
plt.xlabel('y_test')
plt.ylabel('pred')
plt.show()
```

## Data Preprocessing 3

### Using filtered dataset with removed outliers from rm

```python
outl = list(df[((df.medv == 50) & (df.rm > 4) & (df.rm < 8)) | (df.rm < 4) | (df.medv < 30) & (df.rm > 8) | 
               ((df.medv == 50) & (df.lstat > 7))].index)
df3 = df.drop(index=outl)
df_X3 = df3[f_ft]
df_y3 = df3[['medv']]
```

### Comparison

```python
from matplotlib.lines import Line2D

fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(121) 
ax1.scatter(df['rm'].iloc[outl], df['medv'].iloc[outl])
ax1.scatter(df_X3['rm'], df_y3, c='orange')
ax1.set_title('rm vs medv')
ax1.set_xlabel('rm')
ax1.set_ylabel('medv')

ax2 = fig.add_subplot(122) 
ax2.scatter(df['lstat'].iloc[outl], df['medv'].iloc[outl])
ax2.scatter(df_X3['lstat'], df_y3, c='orange')
ax2.set_title('lstat vs medv')
ax2.set_xlabel('lstat')
ax2.set_ylabel('medv')
legend_elements = [Line2D([], [], marker='o', label='Outliers'), 
                    Line2D([], [], marker='o', color='orange', label='Filtered data')]
ax2.legend(handles=legend_elements, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

```python
# splitting
from sklearn.model_selection import train_test_split

X_train3, X_test3, y_train3, y_test3 = train_test_split(df_X3, df_y3, test_size=0.3, random_state=33)
```

```python
# scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train3)
X_train3 = scaler.transform(X_train3)

scaler.fit(X_test3)
X_test3 = scaler.transform(X_test3)
```

## Data Modelling 3

```python
# modelling
from sklearn.linear_model import LinearRegression

boston_model = LinearRegression()
boston_model.fit(X_train3, y_train3)
```

```python
boston_model.score(X = X_train3, y = y_train3)
```

### Model Test
